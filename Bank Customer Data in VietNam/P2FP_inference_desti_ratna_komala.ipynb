## 2- Import Library

```python
# Import Library

import pickle
import joblib
import pandas as pd
import numpy as np
from tensorflow.keras.models import load_model
```

## 3-Load model

```python
load_model = joblib.load("xgboost.pkl")
```

```python
# Load Model

with open('preprocessor_nn.pkl', 'rb') as file_1:
  preprocessor = pickle.load(file_1)

model_nn = load_model('nn_best_model.h5', compile=False)
```

```python
#panggil data post undersampling
test_data=pd.read_csv(r'/content/bank_marketing.csv')
test_dt=test_data.copy()
test_dt
```

## 4- Data inferencing

```python
#Input data baru dari test dan lihat hasil death eventnya, apakah sesuai atau tidak
#data inference
data_inf = {
        'age': 49,
        'job':'management',
        'marital':2,
        'education':'tertiary',
        'default':0,
        'balance':1093,
        'housing': 1,
        'loan':1,
        'duration':243,
        'campaign': 2,
        'pdays':91,
        'last_contact_day': 198,
        'previous': 2
}
data_inf = pd.DataFrame([data_inf])
data_inf
```

## 5-Predict Neural Network

```python
# Preprocessing data inference
data_final = preprocessor.transform(data_inf)
```

```python
# Prediksi data inference

y_inf_pred = np.where(model_nn.predict(data_final) >= 0.5, 1, 0)
y_inf_pred

if y_inf_pred == 0:
  print('Not Subscribe')
else:
  print('Subscribe')
```

## 6-Predict Cat boost

```python
pred_new= load_model.predict(data_inf)
pred_new
```

```python
print(f"Hasil prediksi apakah penumpang puas atau tidak:", end="\n\n")
for _ in pred_new:
    match _:
        case 0:
            print("Not subscribe")
        case 1:
            print("subscribe")
```
